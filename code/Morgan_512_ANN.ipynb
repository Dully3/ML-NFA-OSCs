```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neural_network import MLPRegressor
from joblib import dump,load
from sklearn.metrics import r2_score
from scipy.stats.stats import pearsonr
import shap
import pickle
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
df = pd.read_csv('../data/1586_morgan_512.csv')
df
```

```python
X = df.iloc[:, 5:1029]
Y = df.iloc[:, 1]
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
X_test = X_test.astype('float16')
X_train = X_train.astype('float16')
```

```python
import pickle
with open('../model/Morgan_512_ANN.pkl', 'rb') as f:
    ann_regressor = pickle.load(f)
```

```python
model=ann_regressor
```

```python
####fit and predict###
ann_regressor.fit(X_train, y_train)
y_train_hat = ann_regressor.predict(X_train)
y_test_hat = ann_regressor.predict(X_test)
```

```python
np.savetxt('y_train_Morgan_512.csv', y_train)
np.savetxt('y_test_Morgan_512.csv', y_test)
np.savetxt('y_train_hat_Morgan_512.csv', y_train_hat)
np.savetxt('y_test_hat_Morgan_512.csv', y_test_hat)
```

```python

def R2Score(X,Y):
    r2=metrics.r2_score(X,Y)
    return r2

def PearsonCoefficient(X, Y):
    corr, _ = pearsonr(X,Y)
    return corr
```

```python
PearsonCoefficient(y_test,y_test_hat)
```

```python
R2Score(y_test,y_test_hat)
```

```python
background_distribution = shap.kmeans(X_train,10)
```

```python
explainer = shap.KernelExplainer(model.predict, background_distribution)
```

```python
shap_values = explainer.shap_values(X_test)
```

```python
np.savetxt('shap_values_Morgan_512.csv', shap_values, delimiter=',')
```

```python
feat_names1 = list(df.columns)
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 36  # 字号
plt.rcParams['figure.figsize'] = [4,6]
plt.rcParams['figure.dpi']= 300
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
```

```python
shap.initjs()
# 定义两种颜色
color01 = (143,209,225)
color02 = (15,147,121)
color03 = (46,78,126)
color04 = (53,34,159)

# 将颜色值转换为0-1范围
colors = [tuple([x / 255 for x in color]) for color in [color01, color02, color03, color04]]

# 创建四种颜色之间的渐变
gradient_colors = [colors[0], colors[1], colors[2], colors[3]]


# 生成颜色渐变
n = 256
newcolors = []
for color in gradient_colors:
    rgba_color = cm.colors.to_rgba(color)
    newcolors.append(rgba_color)

newcmp = cm.colors.LinearSegmentedColormap.from_list('BlueYellow', newcolors, N=n)

# 使用修改后的颜色渐变绘制图形
plt.figure()
shap.summary_plot(shap_values,X_test,feature_names = feat_names1,plot_type='dot',max_display=5,show=False)
for fc in plt.gcf().get_children():
    for fcc in fc.get_children():
        if hasattr(fcc, "set_cmap"):
            fcc.set_cmap(newcmp)

plt.title(('Morgan_512_ANN'), fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('../SHAP/SHAP_Morgan_512_ANN_1.tif', dpi = 300)
```
